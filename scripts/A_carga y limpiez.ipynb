### <b><font color="#a9f04d"> Acerca del Dataset </font></b>

Esta recopilación de información, cuidadosamente seleccionada, ofrece una visión integral de la educación en Colombia. Ofrece información importante sobre la variación educativa entre municipios y departamentos. El conjunto de datos abarca diversos aspectos del sistema educativo colombiano, como las tasas de matriculación escolar en niños de 5 a 16 años, el porcentaje de estudiantes que repiten o reprueban un año escolar y la cobertura neta en educación preescolar, primaria, secundaria y media superior. También incluye identificadores geográficos y administrativos, lo que permite un análisis regional detallado. Este conjunto de datos es un recurso valioso para investigadores, educadores y formuladores de políticas públicas. Les ayuda a comprender los desafíos educativos locales y a desarrollar estrategias para mejorar el acceso, la calidad y la equidad educativa en todo el país.

Para el tratamiento de los valores faltantes en el conjunto de datos, se empleó el método IterativeImputer. A diferencia de técnicas más simples como la imputación por media o mediana, IterativeImputer permite estimar los valores ausentes mediante un enfoque iterativo y predictivo, modelando cada variable incompleta como una función del resto. Esto garantiza una mayor coherencia interna entre las variables imputadas y una conservación más realista de la variabilidad de los datos originales. Gracias a este enfoque, se logró reducir el sesgo potencial que podría generarse al eliminar o imputar datos de manera arbitraria, mejorando así la calidad del análisis descriptivo y la robustez de los modelos de predicción utilizados en las etapas posteriores del proyecto.

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

iter_imputer = IterativeImputer()
iter_imputer
```

```python
data_numerico = df_educacion.drop(['ETC', 'DEPARTAMENTO', 'MUNICIPIO'], axis=1)
data_imputed=iter_imputer.fit_transform(data_numerico)
```

```python
pd.DataFrame(data_imputed, columns=data_numerico.columns)
```

```python
# Convertir los datos imputados en un DataFrame
data_imputed_df = pd.DataFrame(data_imputed, columns=data_numerico.columns)

# Concatenar las columnas originales que eliminaste
df = pd.concat([df_educacion[['ETC', 'DEPARTAMENTO', 'MUNICIPIO']], data_imputed_df], axis=1)

# Verificar el resultado
print(df.head())
```

#### <b><font color="#a9f04d"> 1.5.2 Datos en columna "POBLACIÓN_5_16" </font></b>

Nos interesa limpiar la columna "POBLACIÓN_5_16", ya que persenta diversas inconsistencia en los valores respecto al año 2021, son de tipo float cuando deberían ser de tipo entero.

```python
def corregir_poblacion(valor):
    try:
        valor = float(valor)
        if valor < 300:  # probablemente fue escrito como miles con punto decimal
            return int(round(valor * 1000))
        else:
            return int(round(valor))
    except:
        return None  # Por si hay algún valor extraño

# Aplicar la función a la columna
df['POBLACIÓN_5_16'] = df['POBLACIÓN_5_16'].apply(corregir_poblacion)
```
